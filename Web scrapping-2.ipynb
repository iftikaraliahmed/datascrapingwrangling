```python
browser=webdriver.Chrome('chromedriver.exe')
url='https://www.naukri.com/datascience-jobs'
browser.get(url)

```

```python
import pandas as pd
nk=pd.DataFrame()
parents=browser.find_elements_by_css_selector('div.row[type="tuple"]') #tags where <div class='row' type='tuple>'
len(parents)
for job in parents:
    job_id=job.get_attribute('id')
    #print(job_id)
    title=job.find_element_by_css_selector('li.desig').get_attribute('title')
    #print(title)
    company=job.find_element_by_css_selector('span.org').text
    #print(company)
    exp=job.find_element_by_css_selector('span.exp').text
    #print(exp)
    loc=job.find_element_by_css_selector('span.loc').text
    #print(loc)
    key_skills=job.find_element_by_css_selector('span.skill').text
    #print(key_skills)
    sal=job.find_element_by_css_selector('span.salary').text
    #print(sal)
    date=job.find_element_by_css_selector('span.date').text
    #print(date)
    url=job.get_attribute('data-url')
    #print(url)
    curr_row={'Url':url,'Job_ID':job_id,'Title':title,'Company':company,'Experience':exp,'Location':loc,'Key':key_skills,
              'Salary':sal,'Date':date}
    nk=nk.append(curr_row,ignore_index=True)
nk.to_csv('Jobs.csv')
```

```python
browser=webdriver.Chrome('chromedriver.exe')
url='https://www.oyorooms.com/oyos-in-bangalore'
browser.get(url)
```

```python
import pandas as pd
import time

hotel_list=pd.DataFrame()
while True: 
    hotels=browser.find_elements_by_css_selector("div.ember-view.mdl-cell.mdl-cell--12-col.newHotelCard")
    for hotel in hotels:
        try:
            hotel_name=hotel.find_element_by_css_selector('span.newHotelCard__hotelName').text
        except:
            hotel_name=None
        #print(hotel_name)
        try:
            hotel_locations=hotel.find_element_by_css_selector('span.newHotelCard__hotelAddress').text
        except:
            hotel_name=None
        #print(hotel_locations)
        try:
            hotel_rprice=hotel.find_element_by_css_selector('span.newHotelCard__revisedPricing').text
        except:
            hotel_rprice=None
        #print(hotel_rprice)
        try:
            hotel_discount=hotel.find_element_by_css_selector('div.newHotelCard__discount').text
        except:
            hotel_discount=None
        #print(hotel_discount)
        try:
            hotel_rating=hotel.find_element_by_css_selector('span.hotelRating__value.hotelRating__value--verygood').text
        except:
            hotel_rating=None
        #print(hotel_rating)
        try:
            hotel_no_of_rating=hotel.find_element_by_css_selector('div.hotelRating__count').text
        except:
            hotel_no_of_rating=None
        #print(hotel_no_of_rating)
        try:
            hotel_url=hotel.find_element_by_css_selector('img.carousal__image').get_attribute('src')
        except:
            hotel_url=None
        #print(hotel_url)
        curr_row={'Hotel Name:':hotel_name,'Hotel Locations:':hotel_locations,"Hotel Revised Price:":hotel_rprice,
                  "Hotel Discount:":hotel_discount,"Hotel_Discount:":hotel_rating,"Number of ratings:":hotel_no_of_rating,
                  "Hotel link:":hotel_url}
        hotel_list=hotel_list.append(curr_row,ignore_index=True)
hotel_list    
#hotel_list.to_csv('Hotels.csv')    
    
```
