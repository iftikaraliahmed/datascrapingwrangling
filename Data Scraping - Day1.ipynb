# Scrape Paragraphs

```python
#finding the elements or more than one paragraph we are going to extract
ptags = browser.find_elements_by_css_selector('p')
```

```python
len(ptags)
```

```python
ptags_texts = ' '    #to find the length of the paragraphs of the wikipedia datascience #avoid lots of statements
for tag in ptags:
    #print(tag.text)
    text = tag.text
    ptags_texts = ptags_texts + ' ' + text
len(ptags_texts)
```

```python
ptags = browser.find_elements_by_css_selector('p')
for tag in ptags:
    print(tag.text)
    print(tag.get_attribute('style'))
    print(tag.get_attribute('class'))
```

```python
import pandas as pd
df_paras = pd.DataFrame()
ptags = browser.find_elements_by_css_selector('p')
for tag in ptags:
    print(tag.text, tag.get_attribute('class'))
    curr_row = {
    'tag_name':'p', 
    'class': tag.get_attribute('class'), 
    'text': tag.text
    }
    df_paras = df_paras.append(curr_row, ignore_index=True)
df_paras
```

```python
ptags = browser.find_elements_by_css_selector('p.article')  #extract only paragraph texts with the class article. #getting a multiple attribute by specifying class #class selector
for tag in ptags:
    print([tag.text])
```

```python
div_tag = browser.find_element_by_css_selector('div.eda') #Getting a particular attribute inside a parent
div_tag
```

```python
ptags_eda = div_tag.find_elements_by_css_selector('p.article')
[tag.text for tag in ptags_eda]
```

```python
div_tag = browser.find_element_by_css_selector('div[name="ml"]')  #('[name=" "])
ptags_ml = div_tag.find_elements_by_css_selector('p')
[tag.get_attribute('value') for tag in ptags_ml]
```

```python
df1 = pd.DataFrame()
div_tags = browser.find_element_by_css_selector('div.articles')
div_tag = div_tags.find_elements_by_css_selector('div.article')
for class1 in div_tag:
    curr_row = {'title':class1.find_element_by_css_selector('p[name="title"]').text,
                'content': class1.find_element_by_css_selector('p[name="content"]').text
               }
    df1 = df1.append(curr_row, ignore_index = True)
    
df1 = df1.set_index('title')
df1
```

```python
#Alternate way method 
ptags_titles = browser.find_elements_by_css_selector('p[name="title"]')
titles = [tag.text for tag in ptags_titles]

ptags_contents = browser.find_elements_by_css_selector('p[name="content"]')
contents = [tag.text for tag in ptags_contents]

pd.DataFrame({'title': titles,
             'content': contents})
```

```python
div_tag = browser.find_elements_by_css_selector('div.book')
df = pd.DataFrame()
for class1 in div_tag:
    curr_row = {'title':class1.find_element_by_css_selector('li.title').text,
               'author':class1.find_element_by_css_selector('div.author').text,
               'desc':class1.find_element_by_css_selector('div.desc').text,
               'price':class1.find_element_by_css_selector('li[type="price"]').text,
               'category':class1.find_element_by_css_selector('div.category').get_attribute('type')}
    df = df.append(curr_row, ignore_index=True)
df
```
