```python
import numpy as np
import pandas as pd
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
url = 'https://www.gsmarena.com/apple_iphone_6-reviews-6378.php'
browser.get(url)
```

```python
gsm_parent = browser.find_elements_by_css_selector('div.user-thread')#('div[id="user-thread"]')
```

```python
len(gsm_parent)
```

```python
review_df = pd.DataFrame()
for review in gsm_parent:
    try:
        user_name = review.find_element_by_css_selector('li.uname2').text
        review_date = review.find_element_by_css_selector('li.upost').text
        review_text = review.find_element_by_css_selector('p.uopin').text
        rating_value = review.find_element_by_css_selector('p.uopin').text
        try:
            up_votes = review.find_element_by_css_selector('span.thumbs-score').text
        except:
            up_votes = None
        curr_row = {
            'User Name':user_name,
            'Review Date':review_date,
            'Review Text':review_text,
            'Rating Value':rating_value,
            'Up votes':up_votes
        }
    except:
        user_name = None
        review_date = None
        review_text = None
        rating_value = None

    review_df = review_df.append(curr_row,ignore_index = True)
review_df
```
