```python
from selenium import webdriver
```

```python
import pandas as pd
import numpy as np
import os
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
url = 'https://www.naukri.com/data-science-jobs'
browser.get(url)
```

```python
parents = browser.find_elements_by_css_selector('div.row[type="tuple"]')
len(parents)
```

```python
df=pd.DataFrame()
for job in parents:
    job_id = job.get_attribute('id')
    job_url = job.get_attribute('data-url')
    job_title = job.find_element_by_css_selector('li.desig').get_attribute('title')
    company = job.find_element_by_css_selector('span.org').text
    experience = job.find_element_by_css_selector('span.exp').text
    location = job.find_element_by_css_selector('span.loc').text
    key_skill = job.find_element_by_css_selector('span.skill').text
    salary = job.find_element_by_css_selector('span.salary').text
    date = job.find_element_by_css_selector('span.date').text
    curr_row = {'Job_url':job_url, 'Job_id':job_id, 'Job_title':job_title,'Company':company,'Experience':experience,'Location':location,'Key_skill':key_skill, 'Salary':salary, 'Date':date}
    
    
    df=df.append(curr_row, ignore_index=True)
    
```

```python
df = df[['Job_url','Job_id','Job_title','Company','Key_skill','Location','Salary','Date']]
df
```

```python
df.to_csv('jobs_datascience.csv', index=False)
```
