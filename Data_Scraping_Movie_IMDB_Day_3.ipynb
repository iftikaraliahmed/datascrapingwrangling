```python
col_names = [th.text for th in th_tags]
col_names
```

# Get data

```python
col_names = [th.text for th in th_tags]
col_names[0] = 'Poster'
col_names[-1] = 'Watchlist'
imdb =pd.DataFrame(columns=col_names)
tr_tags=table_tag.find_elements_by_css_selector('tbody > tr')
```

```python
len(tr_tags)
```

```python
for tr in tr_tags:
    td_tags = tr.find_elements_by_css_selector('td')
    #print(len(td_tags))
    td_data = [td.text for td in td_tags]
    row_series = pd.Series(td_data, index=col_names)
    imdb = imdb.append(row_series, ignore_index=True)
```

```python
imdb.head()
```

```python
col_names = [th.text for th in th_tags]
col_names[0] = 'Poster'
col_names[-1] = 'Watchlist'
imdb =pd.DataFrame(columns=col_names)
tr_tags=table_tag.find_elements_by_css_selector('tbody > tr')

for tr in tr_tags:
    td_tags = tr.find_elements_by_css_selector('td')
    td_data = [td.text for td in td_tags]
    
    img_src = td_tags[0].find_element_by_css_selector('img')
    img_src = img_src.get_attribute('src')
    td_data[0] = img_src
    
    row_series = pd.Series(td_data, index=col_names)
    imdb = imdb.append(row_series, ignore_index=True)
imdb.head()
```

```python
import glob  #to list out all the files in current working directory
glob.glob('*')
```

```python
#saving image from url
import urllib                            #library to retrieve url's
import os
img_url = imdb.loc[0, 'Poster']
movie_name = imdb.loc[0, 'Rank & Title']  
img_extn = img_url[img_url.rfind('.'):]
folder_name = 'movies_images_imdb'

if not folder_name in glob.glob('*'):     #checks whether folder name is existed or not
    os.makedirs(folder_name)
img_name = folder_name+ '/' + movie_name + img_extn


print(img_url, img_name)                        #movie_name
urllib.request.urlretrieve(img_url,img_name)    #movie_name
```
